```python
from numpy import linspace, cos, sin, tanh
import matplotlib.pyplot as plt
%matplotlib inline

x = linspace(-6, 6, 50)
y = cos(x)
z = sin(x)
t = tanh(x)

fig = plt.figure()
ax1 = plt.subplot()
ax2 = plt.subplot()
ax3 = plt.subplot()

ax1.scatter(x, y, label='sin')
ax2.scatter(x, z, label='cos')
ax3.scatter(x, t, label='tanh')
plt.legend()
```

```python
# import the libraries we're using
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from scipy import stats
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/acc_a/Documents/GitHub/Regression_techniques/ames_housing/train.csv')
df.head()
```

```python
#histogram
sns.distplot(df['SalePrice'])
ax1 =plt.subplot() 
ax1.set(xlabel='saleprice', ylabel='density')
```

```python
#box plot overallqual/saleprice

data = pd.concat([df['SalePrice'], df['OverallQual']], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x='OverallQual', y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
```

```python
#correlation matrix
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

```python
# sort correlation from highest to lowest
corr = df.select_dtypes(include = ['float64', 'int64']).iloc[:, :].corr()
cor_dict = corr['SalePrice'].to_dict()
del cor_dict['SalePrice']
print("List the features in descending order of their correlation with SalePrice:\n")
for ele in sorted(cor_dict.items(), key=lambda x: -abs(x[1])):
    print("{0}:   {1}".format(*ele))
```

```python
#scatterplot
sns.set()
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', '1stFlrSF']
sns.pairplot(df[cols], size = 2.5, kind='reg') #fit linear regression
plt.show();
```
